```python
# Centralizing the letters.

def Centralize_Image(intput_path, output_path):
    
    Images_Names = os.listdir(input_path)
    
    for Image_Name in Images_Names:
        
        os.chdir(input_path)
        Original_Image = cv2.imread(Image_Name,0)
        mat = Original_Image

        # Finding the Starting and Ending Pixel using searching by raws.

        m = mat.shape[0]       # number of rows
        n = mat.shape[1]       # number of columns

        count = 0
        start_r = (0, 0)
        end_r = (28, 28)

        for i in range(m):
            for j in range (n):
                if mat[i][j] < 255:
                    count += 1
                    if count == 1:
                        start_r = (i, j)
                    end_r = (i,j)

        count = 0
        start_c = (0, 0)
        end_c = (28, 28)

        for j in range(n):
            for i in range (m):
                if mat[i][j] < 255:
                    count += 1
                    if count == 1:
                        start_c = (i, j)
                    end_c = (i, j)

        start = (start_r[0], start_c[1])
        end = (end_r[0], end_c[1])

        Image_Center = (25, 25)

        m_text = end[0] - start[0] + 1
        n_text = end[1] - start[1] + 1

        new_start = (int(Image_Center[0] - (m_text - 1)/2), int(Image_Center[1] - n_text/2)) 
        new_end = (int(Image_Center[0] + (m_text - 1)/2), int(Image_Center[1] + n_text/2 - 1))

        help_mat = [] 
        for i in range(start[0], end[0] + 1):
            for j in range(start[1], end[1] + 1):
                help_mat.append(mat[i][j])

        for i in range(start[0], end[0] + 1):
            for j in range(start[1], end[1] + 1):
                mat[i][j] = 255

        k = 0
        for i in range(new_start[0], new_end[0] + 1):
            for j in range(new_start[1], new_end[1] + 1):
                mat[i][j] = help_mat[k]  
                k += 1

        os.chdir(output_path)        
        cv2.imwrite(Image_Name,mat) 
```

```python
# Small_Small_Letter_Pairs

from PIL import Image, ImageDraw, ImageFont

# Function to generate pairs of small letters
def generate_letter_pairs():
    pairs = []
    for i in range(97, 123):  # ASCII range for lowercase letters
        for j in range(97, 123):
            pairs.append(chr(i) + chr(j))
    return pairs

# Function to arrange pairs in a grid
def arrange_in_grid(pairs, columns=26):
    rows = [pairs[i:i+columns] for i in range(0, len(pairs), columns)]
    return rows

# Function to create and save the image
def save_image(grid, filename, font_path, output_folder):
    # Calculate image dimensions based on grid size
    width = len(grid[0]) * 80
    height = len(grid) * 80

    # Create white image
    img = Image.new('RGB', (width, height), color='white')
    draw = ImageDraw.Draw(img)

    # Increase font size
    font_size = 40

    # Iterate through each font file in the folder
    for font_file in os.listdir(font_path):
        font = ImageFont.truetype(os.path.join(font_path, font_file), font_size)

        # Draw letters in black with improved anti-aliasing
        x_offset = 0
        y_offset = 0
        for row in grid:
            for pair in row:
                text_bbox = draw.textbbox((0, 0), pair, font=font)
                text_width = text_bbox[2] - text_bbox[0]
                text_height = text_bbox[3] - text_bbox[1]
                draw.text(((80 - text_width) / 2 + x_offset, (80 - text_height) / 2 + y_offset), pair, fill='black', font=font, align='center')
                x_offset += 80
            y_offset += 80
            x_offset = 0

        # Save the image with high quality
        output_filename = os.path.join(output_folder, f"{font_file.split('.')[0]}.png")
        img.save(output_filename, quality=95)
        img = Image.new('RGB', (width, height), color='white')
        draw = ImageDraw.Draw(img)

if __name__ == "__main__":
    font_path = r"C:\Users\Mohamad Hallak\Desktop\PhD\Datasets\HWR\Font_Families"
    output_folder = r"C:\Users\Mohamad Hallak\Desktop\PhD\Datasets\HWR\Images\Small_Small_Letter_Pairs"
    pairs = generate_letter_pairs()
    grid = arrange_in_grid(pairs, columns=26)

    # Save images using fonts from the specified folder
    save_image(grid, ".png", font_path, output_folder)
```

```python
# Segmenting the characters

'''
m: the number of rows of the image.
n: the number of columns of the image.
image : the source image.
n_Lines: the number of extracted lines

The extracted characters will be stored as png images

'''
def Character_Segmentation(input_path, output_path, i):
    global c
    os.chdir(input_path)
    
    image = cv2.imread(str(i) + '.png', 0)

    m = image.shape[0]
    n = image.shape[1]


    columns_sum = []
    for j in range(n):
        columns_sum.append(np.sum(image[:, j]))

    max_sum = max(columns_sum)  
    Class = []

    for j in range(n):
        if columns_sum[j] == max_sum:
            Class.append(0)
            Counter = max(Class)
        else: 
            if j == 0:
                Counter = 0
            Class.append(Counter + 1)


    n_Characters = max(Class)


    for character_num in range(1, n_Characters + 1):

        Count = 0
        for j in range(n):
            if Class[j] == character_num:
                Count += 1
                if Count == 1:
                    Char_Start = j
                Char_End = j   

        os.chdir(output_path)        
        Character = image[:, Char_Start:Char_End + 1]
        c += 1
        cv2.imwrite(str(c) +'.png',Character)
```
